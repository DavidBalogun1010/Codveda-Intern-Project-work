```python
## Importing libraries 

import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 

sns.set_style("whitegrid")
```

```python
### Loading the dataset into a dataframe and Exploring the dataset

df = pd.read_csv("datasets\stock price.csv")
```

```python
### Displaying the 5 rows
df.head()
```

```python
### displaying the dataset shape i.e number of rows and columns
df.shape
```

```python
#### Exploring the dataset types
df.info()
```

-- From the datatypes, we discovered that `date` column is object. we'll have to change it's type to a date type.

```python
### Converting the date column into a proper date type for an object type.
df['date'] = pd.to_datetime(df['date'],errors ="coerce")
```

```python
df.info()
```

```python
#### Checking for Missing values in the dataset
df.isnull().sum()
```

```python
df.isnull().sum() / df.shape[0] * 100
```

From the result, we can see that we have some missing values in OPEN,HIGH and LOW column in the dataset. In order, to resolve this. we can drop the columns of the missing value because the missing is sigificantly small with a percentage of 0.002,0.001 and 0.001 percent of overall dataset.

```python
### Handling the Missing values

df.dropna(inplace=True)
```
