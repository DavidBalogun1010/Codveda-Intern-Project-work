### Calculating RSI, MACD , SMA on the close price for each stock
-- Feature Engineering: We'll be considering some technical time series stock indicators to help US further understand the dataset.

```python
from ta.momentum import RSIIndicator
from ta.trend import MACD,SMAIndicator
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python

XOM_stock.head()
```

RSI : Compares the magnitude of recent gains and losses over a specified time period to measure speed and change of price movements of a security. It is primarily used to attempt to identify overbought or oversold conditions in the trading of an asset

Moving Average Convergence Divergence (MACD) :Is a trend-following momentum indicator that shows the relationship between two moving averages of prices.

SMA - Simple Moving Average

```python
XOM = XOM_stock.copy()
```

```python
XOM['rsi'] = RSIIndicator(XOM['close'],window =14).rsi()
```

```python
macd = MACD(XOM['close'])
XOM['macd'] = macd.macd()
XOM['macd_signal'] = macd.macd_signal()
XOM['macd_diff'] = macd.macd_diff() 
XOM['sma_20'] = SMAIndicator(XOM['close'],window=20).sma_indicator()
XOM['sma_50'] = SMAIndicator(XOM['close'],window=50).sma_indicator()
```

```python
XOM.head(10)
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report   
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier 


RANDOM_STATE = 42
TARGET_HORIZON = 1
```

```python
XOM['return'] = XOM['close'].pct_change()
```

```python
XOM.head()
```

```python
### Lag feature creation
XOM['close_lag1'] = XOM['close'].shift(1)
XOM['close_lag5'] = XOM['close'].shift(5)

XOM['RSI_lag1'] = XOM['rsi'].shift(1)
XOM['volume_lag3'] = XOM['volume'].shift(3)

XOM['volatility_10'] = XOM['return'].rolling(window=10).std()
```

```python
XOM.head()
```

```python
#### Target Creation
XOM['target'] = (XOM['close'].shift(-TARGET_HORIZON) > XOM['close']).astype(int)
```

```python
XOM.shape
```

### Drop the NaNs after all engineering 

```python
XOM = XOM.dropna().reset_index(drop=True)
```

```python
#### Time series cross validation
tscv = TimeSeriesSplit(n_splits=5)

results = {}

for name, pips in pipeline.items():


    scores = []

    for train_idx, val_idx in tscv.split(X_train):
        X_train_fold, X_val_fold = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_train_fold, y_val_fold = y_train.iloc[train_idx], y_train.iloc[val_idx]
        
        pips.fit(X_train_fold, y_train_fold)
        val_preds = pips.predict(X_val_fold)
        score = accuracy_score(y_val_fold, val_preds)
        scores.append(score)
    results[name] = sum(scores) / len(scores)

print("cross validation result: ", results)
```

## Using Facebook Prophet to predict the market price.

```python
from prophet import Prophet 
```

```python
prophet_df['close'] = prophet_df['close'].astype('float')
```

```python
prophet_df = prophet_df.rename(columns = 
    {"date":'ds',
     "close":'y'
     })
```

```python
split_n = int(len(prophet_df) * 0.8)
```

```python
train, test = prophet_df.iloc[:split_n], prophet_df.iloc[split_n:]
```

```python
### Building the Model

model = Prophet()
model.fit(train)
```
